## Cleaning and Merging of Datasets 

```python
#dropping unnecessary address columns from train data since 
#we are given information on the lat. and long. of each row 
train.drop(columns=['Address', 'Block', 'Street', 'AddressNumberAndStreet', 'AddressAccuracy'], inplace=True)
train.shape
```

```python
# check for duplicates
train.duplicated().sum()
```

```python
# remove for duplicates
train = train.drop_duplicates()
train.shape
```

```python
# check for duplicates
test.duplicated().sum()
```

```python
#dropping unnecessary address columns from test data since 
# we are given information on the lat. and long. of each row 
test.drop(columns=['Address', 'Block', 'Street', 'AddressNumberAndStreet', 'AddressAccuracy'], inplace=True)
test.shape

```

```python
#merging of both dataset
combined = pd.concat([train, test])
combined.shape
```

```python
#lowercase all columns 
combined.columns = combined.columns.map(lambda x: x.lower())
```

```python
#converting date to day, month and year
combined.date = pd.to_datetime(combined.date)
combined['month']=combined['date'].apply(lambda x: x.to_pydatetime().month)
combined['year']=combined['date'].apply(lambda x: x.to_pydatetime().year)
combined['day_of_week']=combined['date'].apply(lambda x: x.to_pydatetime().weekday())
combined['day']=combined['date'].apply(lambda x: x.to_pydatetime().day)
```

```python
#converting categorical variables to dummies
combined=pd.get_dummies(combined,columns=['species', 'trap'])
```

```python
combined.head()
```

```python
combined.info()
```

# Seperate Train and Test Sets 

```python
# Differentiating test and train dataset
train = combined.loc[combined['wnvpresent'] >= 0]
test = combined.loc[combined['wnvpresent'].isnull()]
```

```python
# Dropping columns 'nummosquitos' and 'wnvpresent' columns from test datasets and id from train dataset
# since it is redundant
test.drop(['nummosquitos', 'wnvpresent'], axis = 1, inplace=True)
train.drop(['id'], axis = 1, inplace=True)
```

```python
# Double-checking on our train and test datasets
print(train.shape, test.shape)
```

## Saving our cleaned train and test dataset

```python
train.to_csv('../Assets/train_cleaned.csv', index=False)
test.to_csv('../Assets/test_cleaned.csv', index=False)
```

## Visualisation on each month of which has the most mosquitoes 

```python
plt.subplots(figsize=(20,10))
plt.subplot(1,2,1),sns.barplot('month','nummosquitos',data=train,hue='wnvpresent')  
ax2=plt.subplot(1,2,2),sns.barplot('year','nummosquitos',data=train,hue='wnvpresent')
```

### Comments: 
First case of WNV discovered in Chicago was July 2007. <br>
WNV began subsiding after July. <br>
There was a massive drop in cases in the month of November 2007. <hr>

```python
def wnv_year():
    df = train
    years = [2007,2009,2011,2013]
    fig,ax = plt.subplots(nrows=1,ncols=len(years), sharey=True)
    fig.set_size_inches(16, 5)
    for i, year in enumerate(years):
        wnv_0= df[(df['year']==years[i])&(df['wnvpresent']==0)]
        wnvn_1= df[(df['year']==years[i])&(df['wnvpresent']==1)]
        ax[i].set_title(year)
        sns.regplot('longitude','latitude',data=wnv_0,ax=ax[i],color='steelblue', fit_reg=False)
        sns.regplot('longitude','latitude',data=wnvn_1,ax=ax[i],color='orange', fit_reg=False)
```

```python
#Visualisaton of location of virus from 2007 to 2013
wnv_year()
```

```python
def wnv_year_month(wnv_year):
    df = train
    months = ['Jun','July', 'Aug','Sept','Oct', 'Nov']
    fig,ax = plt.subplots(nrows=1,ncols=len(months), sharey=True)
    fig.set_size_inches(16, 5)
    for i, month in enumerate(months):
        wnv_0= df[(df['year']==wnv_year)&(df['wnvpresent']==0)&(df['month']==i+5)]
        wnvn_1= df[(df['year']==wnv_year)&(df['wnvpresent']==1)&(df['month']==i+5)]
        ax[i].set_title(month)
        sns.regplot('longitude','latitude',data=wnv_0,ax=ax[i],color='steelblue', fit_reg=False)
        sns.regplot('longitude','latitude',data=wnvn_1,ax=ax[i],color='orange',fit_reg=False)
```

```python
#visualising spread of virus across months in 2007
wnv_year_month(wnv_year = 2007)
```

```python
#visualising spread of virus across months in 2009
wnv_year_month(wnv_year = 2009)
```

```python
#visualising spread of virus across months in 2011
wnv_year_month(wnv_year = 2011)
```

```python
#visualising spread of virus across months in 2013
wnv_year_month(wnv_year = 2013)
```
