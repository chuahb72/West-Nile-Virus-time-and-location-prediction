```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    accuracy_score, 
    recall_score,
    precision_score,
    roc_auc_score,
    make_scorer, 
    roc_curve, 
    precision_recall_curve, 
    auc
)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
plt.style.use('fivethirtyeight')
```

```python
# read cleaned train,test and weather files

train_cleaned = pd.read_csv('../Assets/train_cleaned.csv')
test_cleaned = pd.read_csv('../Assets/test_cleaned.csv')
weather_cleaned = pd.read_csv('../Assets/weather_cleaned.csv')
```

```python
weather_cleaned.shape
```

```python
train_cleaned.shape
```

```python
# Average the values of station 1 and station 2 
# Remove station column 
weather_cleaned = weather_cleaned.groupby('Date').mean().drop(columns='Station')
```

```python
# Drop engineered columns with Year, Month and Day, as it is already in the train file
weather_cleaned.reset_index(inplace=True)
weather_cleaned.drop(columns=['Year','Month','Day'],inplace=True)
weather_cleaned.head()
```

```python
# Merge train file with weather file
train_merge = pd.merge(train_cleaned, weather_cleaned,how='left', left_on='date', right_on='Date')
train_merge.shape
```

```python
# Merge test file with weather file, has 1 less column than train (target variable: wnvpresent)
test_merge = pd.merge(test_cleaned, weather_cleaned,how='left', left_on='date', right_on='Date')
test_merge.shape
```

```python
# Remove nummoquitos as it is not in test file 
# remove date and Date because it is already represented by Year, Month, Week and Day
train_merge.drop(columns=['date','Date','nummosquitos'],inplace=True)
train_merge.head()
```

```python
test_merge.drop(columns=['date','Date'],inplace=True)
test_merge.head()
```

```python
# Set target variable y = 'wnvpresent', the rest as X: predictor features
target = 'wnvpresent'
features = train_merge.drop(columns=target).columns
X = train_merge[features]
y = train_merge[target]
y = y.astype(int)
```

```python
# Train test split, stratify according to y
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y)
```

```python
# Train test split 3:1 ratio
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Initiating models:
ada = AdaBoostClassifier(base_estimator=DecisionTreeClassifier())
gbc = GradientBoostingClassifier()
```

```python
# Initiating models with pipeline:
# StandardScaler is applied in order to interpret the effect of the predictor variables on the same scale
ada_pipe = Pipeline([
    ('ss', StandardScaler()),
    ('ada', ada)
])
```

```python
# StandardScaler is applied in order to interpret the effect of the predictor variables on the same scale
gbc_pipe = Pipeline([
    ('ss', StandardScaler()),
    ('gbc', gbc)
])
```

```python
scorer = roc_auc_score
```

```python
# Here we iterate through various parameters for both algorithms to find the 
# best performing parameters of a model

scoring = make_scorer(scorer)
ada_param_distributions = {
    'ada__n_estimators': [50,100],
    'ada__base_estimator__max_depth': [1,2],
    'ada__learning_rate': [.9, 1.]
}

gbc_param_distributions = {
    'gbc__max_depth': [2,3,4],
    'gbc__n_estimators': [100, 125, 150],
    'gbc__learning_rate': [.08, .1, .12]
}
```

```python
# Conduct randomized cross validation on the pipelines, searching through the pre-established parameters above
ada_rs = RandomizedSearchCV(
    ada_pipe, param_distributions=ada_param_distributions,
    scoring=scoring,
    n_jobs=-1,
    cv=5,
    verbose=1
)


gbc_rs = RandomizedSearchCV(
    gbc_pipe, param_distributions=gbc_param_distributions,
    scoring=scoring,
    n_jobs=-1,
    cv=5,
    verbose=1
)
```

```python
# Get metrics for 3 different models for one dataset
models = {'ADA Boost': ada_rs,'Gradient Boost': gbc_rs}
score_df = pd.DataFrame()                                   #Create a blank dataframe to concat iterations later
cols = []
for key, model in models.items():
    cols.append(key)
    model.fit(X_train, y_train)                         #Run each model through fitting, predicting and scoring
    y_pred = model.predict(X_test)  
    y_score = model.predict_proba(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    auc_roc = roc_auc_score(y_test, y_score[:, 1])
    data = {
        'Accuracy': accuracy,
        'Recall': recall, 
        'Precision': precision,
        'ROC AUC': auc_roc,
        'Model': model.best_estimator_ # Included to easily call the model via masking if required
    }
    col = pd.Series(data)           # Create a pandas series with the scores so the scores have an index assigned
    df = pd.DataFrame(col)          # Change the series into a DataFrame  
    
    score_df = pd.concat([score_df, df], axis=1) # Concatenate the dataframe of scores with each iteration
score_df.columns = cols              # Name the columns using model name

# Print dataframe without model
metrics = [metric for metric in score_df.index if metric != 'Model']
score_df.loc[metrics, :]
```

```python
# Automatically calls the model based on highest ROC AUC score
scores = score_df.loc['ROC AUC', :]
win_model = scores[scores == scores.max()]
win_model = win_model.index
win_model = win_model.format()
win_model = win_model[0]
print('Winning Model: {}'.format(win_model))
model = score_df.loc['Model', win_model]
```

```python
y_pred = model.predict(X_test)
y_score = model.predict_proba(X_test)
fpr, tpr, thresholds = roc_curve(y_test, y_score[:, 1])
precision, recall, thresholds = precision_recall_curve(y_test, y_score[:,1])
```

```python
print("roc_auc_score: ", round(roc_auc_score(y_test, y_score[:, 1]), 3))
print("precision-recall_auc_score: ", round(auc(recall,precision), 3))
plt.subplots(figsize=(16, 3))
plt.subplot(1, 3, 1)
xx = np.linspace(0, 1, 1000)
yy = xx
plt.plot(xx, yy, '--')
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

plt.subplot(1, 3, 2)
plt.plot([0, 1], [0.01, 0.01], '--')
plt.plot(recall, precision, color='orange')
plt.xlabel('recall')
plt.ylabel('precision')

plt.subplot(1, 3, 3)
plt.plot(thresholds, precision[:-1], 'gold', label="Precision")
plt.plot(thresholds, recall[:-1], 'purple', label='Recall')
plt.xlabel('thresholds')
plt.ylabel('Score')
plt.legend(loc='best');
```

```python
threshold = 0.01
```

```python
if win_model == 'ADA Boost':
    most = model.named_steps['ada'].feature_importances_
if win_model == 'Gradient Boost':
     most = model.named_steps['gbc'].feature_importances_
```

```python
important_feature = pd.DataFrame({'Feature': X_train.columns, 'Importance': most})
important_feature = important_feature.sort_values('Importance', ascending=True)
important_feature = important_feature[important_feature['Importance'] > threshold]
important_feature.plot.barh('Feature', 'Importance', figsize=(20, 15))
plt.title('Features Importance');
```
