# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier

from sklearn.metrics import (
    accuracy_score, 
    recall_score,
    precision_score,
    roc_auc_score,
    classification_report,
    confusion_matrix
)

pd.set_option("display.max.columns", None)
plt.style.use('fivethirtyeight')
```

# Read datasets

```python
train_cleaned = pd.read_csv('../Assets/train_cleaned.csv')  # training set
test_cleaned = pd.read_csv('../Assets/test_cleaned.csv')  # kaggle test set
weather_cleaned = pd.read_csv('../Assets/weather_cleaned.csv')  # wearher data
```

```python
# check dataset shapes
print(train_cleaned.shape)
print(test_cleaned.shape)
print(weather_cleaned.shape)
```

```python
# Average the values of station 1 and station 2 
weather_cleaned = weather_cleaned.groupby('Date').mean().drop(columns='Station')
```

```python
# Drop engineered columns with Year, Month and Day
weather_cleaned.reset_index(inplace=True)
weather_cleaned.drop(columns=['Year','Month','Day'],inplace=True)
weather_cleaned.head()
```

```python
weather_cleaned.shape
```

```python
# merge train sets to weather data
train_merge = pd.merge(train_cleaned, weather_cleaned,how='left', left_on='date', right_on='Date')
train_merge.shape
```

```python
train_merge.drop(columns=['date','Date','nummosquitos'],inplace=True)
train_merge.head()
```

```python
# merge kaggle test sets to weather data
test_merge = pd.merge(test_cleaned, weather_cleaned,how='left', left_on='date', right_on='Date')
test_merge.shape
```

```python
test_merge.drop(columns=['date','Date','id'],inplace=True)
test_merge.head()
```

```python
test_merge.shape
```

```python
test_merge.isnull().sum().max()
```

# train and validation sets split

```python
target = 'wnvpresent'
features = train_merge.drop(columns=target).columns
X = train_merge[features]
y = train_merge[target]
y = y.astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Baseline Score

We are dealing with a very unbalanced dataset<br>
With only 5% 'positive' class, Accuracy will not be meaningful for model evaluation<br>
It's more important to predict as many TRUE WNV as possible since there are more grave health and medical cost concerns<br>
Therefore Type 2 errors (real wnv cases wrongly classified as non case) is a more serious error to avoid<br>
We need have more considerations for Sensitivity (recall) scores.<br>
<br>
Objective is to FIRST create a model that can generalize well followed by best ROC-AUC and Recall scores<br>

```python
y_test.value_counts(normalize=True)
```

### Scale all features to same order

```python
ss = StandardScaler()
X_train_ss = ss.fit_transform(X_train)
X_test_ss = ss.transform(X_test)
X_ss = ss.transform(X)
kaggle_test_ss = ss.transform(test_merge)
```

# Logistic Regression

```python
# Gridsearch parameters
logr_params = {
    'penalty':['l1', 'l2'],
    'C':np.logspace(-4,4),
    'class_weight': [None, 'balanced']}

# Score base on roc_auc
logr_gridsearch = RandomizedSearchCV(LogisticRegression(solver='liblinear'), 
                              logr_params,
                              scoring='roc_auc',
                              cv=5, 
                              verbose=1, 
                              n_jobs=-1)

logr_gridsearch = logr_gridsearch.fit(X_train_ss, y_train)
```

```python
print(f"cv best score mean: {round(logr_gridsearch.best_score_, 4)}")
print(f"cv best score std: {round(logr_gridsearch.cv_results_['std_test_score'][logr_gridsearch.best_index_], 4)}")
print(f"cv best param: {logr_gridsearch.best_params_}")
print("")
print(f"train set score: {round(logr_gridsearch.score(X_train_ss, y_train), 4)}")
print(f"validation set score: {round(logr_gridsearch.score(X_test_ss, y_test), 4)}")
print("")
pred = logr_gridsearch.predict(X_test_ss)
pred_proba = [i[1] for i in logr_gridsearch.predict_proba(X_test_ss)]
pred_df = pd.DataFrame({'true_values': y_test,
                        'pred_probs':pred_proba})

print(f"ROC AUC score :{round(roc_auc_score(pred_df['true_values'], pred_df['pred_probs']), 4)}")
print(f"Recall score :{round(recall_score(y_test, pred), 4)}")
print(f"Precision score :{round(precision_score(y_test, pred), 4)}")
print(f"Accuracy score :{round(accuracy_score(y_test, pred), 4)}")
```

# Random Forest

```python
rf_params = {
    'n_estimators': list(range(400, 601)),
    'max_features' : list(range(30, 50)),
    'max_depth': list(range(2, 10)),
    'min_samples_split': list(range(2, 9)),
    'class_weight': [None, 'balanced', 'balanced_subsample']
}

rf_gridsearch = RandomizedSearchCV(RandomForestClassifier(), 
                              rf_params,
                              scoring='roc_auc',
                              cv=5, 
                              verbose=1, 
                              n_jobs=-1)

rf_gridsearch = rf_gridsearch.fit(X_train_ss, y_train)
```

```python
print(f"cv best score mean: {round(rf_gridsearch.best_score_, 4)}")
print(f"cv best score std: {round(rf_gridsearch.cv_results_['std_test_score'][rf_gridsearch.best_index_], 4)}")
print(f"cv best param: {rf_gridsearch.best_params_}")
print("")
print(f"train set score: {round(rf_gridsearch.score(X_train_ss, y_train), 4)}")
print(f"validation set score: {round(rf_gridsearch.score(X_test_ss, y_test), 4)}")
print("")
pred = rf_gridsearch.predict(X_test_ss)
pred_proba = [i[1] for i in rf_gridsearch.predict_proba(X_test_ss)]
pred_df = pd.DataFrame({'true_values': y_test,
                        'pred_probs':pred_proba})

print(f"ROC AUC score :{round(roc_auc_score(pred_df['true_values'], pred_df['pred_probs']), 4)}")
print(f"Recall score :{round(recall_score(y_test, pred), 4)}")
print(f"Precision score :{round(precision_score(y_test, pred), 4)}")
print(f"Accuracy score :{round(accuracy_score(y_test, pred), 4)}")
```

# Extremely Randomized Trees

```python
et_params = {
    'n_estimators': list(range(400, 601)),
    'max_features' : list(range(20, 40)),
    'max_depth': list(range(5, 16)),
    'min_samples_split': list(range(2, 9)),
    'class_weight': [None, 'balanced', 'balanced_subsample']
}

et_gridsearch = RandomizedSearchCV(ExtraTreesClassifier(), 
                              et_params,
                              scoring='roc_auc',
                              cv=5, 
                              verbose=1, 
                              n_jobs=-1)

et_gridsearch = et_gridsearch.fit(X_train_ss, y_train)
```

```python
print(f"cv best score mean: {round(et_gridsearch.best_score_, 4)}")
print(f"cv best score std: {round(et_gridsearch.cv_results_['std_test_score'][et_gridsearch.best_index_], 4)}")
print(f"cv best param: {et_gridsearch.best_params_}")
print("")
print(f"train set score: {round(et_gridsearch.score(X_train_ss, y_train), 4)}")
print(f"validation set score: {round(et_gridsearch.score(X_test_ss, y_test), 4)}")
print("")
pred = et_gridsearch.predict(X_test_ss)
pred_proba = [i[1] for i in et_gridsearch.predict_proba(X_test_ss)]
pred_df = pd.DataFrame({'true_values': y_test,
                        'pred_probs':pred_proba})

print(f"ROC AUC score :{round(roc_auc_score(pred_df['true_values'], pred_df['pred_probs']), 4)}")
print(f"Recall score :{round(recall_score(y_test, pred), 4)}")
print(f"Precision score :{round(precision_score(y_test, pred), 4)}")
print(f"Accuracy score :{round(accuracy_score(y_test, pred), 4)}")
```

```python
# Create figure.
plt.figure(figsize = (10,10))

# Create threshold values.
thresholds = np.linspace(0, 1, 200)

# Define function to calculate sensitivity. (True positive rate.)
def TPR(df, true_col, pred_prob_col, threshold):
    true_positive = df[(df[true_col] == 1) & (df[pred_prob_col] >= threshold)].shape[0]
    false_negative = df[(df[true_col] == 1) & (df[pred_prob_col] < threshold)].shape[0]
    return true_positive / (true_positive + false_negative)
    

# Define function to calculate 1 - specificity. (False positive rate.)
def FPR(df, true_col, pred_prob_col, threshold):
    true_negative = df[(df[true_col] == 0) & (df[pred_prob_col] <= threshold)].shape[0]
    false_positive = df[(df[true_col] == 0) & (df[pred_prob_col] > threshold)].shape[0]
    return 1 - (true_negative / (true_negative + false_positive))
    
# Calculate sensitivity & 1-specificity for each threshold between 0 and 1.
tpr_values = [TPR(logr_pred_df, 'true_values', 'pred_probs', prob) for prob in thresholds]
fpr_values = [FPR(logr_pred_df, 'true_values', 'pred_probs', prob) for prob in thresholds]
 
# Plot ROC curve.
plt.plot(fpr_values, # False Positive Rate on X-axis
         tpr_values, # True Positive Rate on Y-axis
         label='ROC Curve')

# Plot baseline. (Perfect overlap between the two populations.)
plt.plot(np.linspace(0, 1, 200),
         np.linspace(0, 1, 200),
         label='baseline',
         linestyle='--')

# Label axes.
plt.title('Receiver Operating Characteristic Curve', fontsize=22)
plt.ylabel('Sensitivity', fontsize=18)
plt.xlabel('1 - Specificity', fontsize=18)

# Create legend.
plt.legend(fontsize=16);
```
